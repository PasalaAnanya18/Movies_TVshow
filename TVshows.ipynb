```python
import requests
import pandas as pd
```

```python
api_key='acad35ac31cf165a131b32e433a3bd66'
url = f'https://api.themoviedb.org/3/tv/popular?api_key={api_key}&language=en-US&page=1'
response=requests.get(url)
data=response.json()
```

```python
data['results']
```

```python
df=pd.DataFrame(data['results'])
```

```python
df=df[['id','adult','overview','popularity','poster_path','name','vote_average','genre_ids','vote_count']]
```

```python
df.head()
```

```python
for i in range(1,501):
  response=requests.get(f'https://api.themoviedb.org/3/tv/popular?api_key={api_key}&language=en-US&page={i}')
  df1=pd.DataFrame(response.json()['results'])
  df1=df1[['id','adult','overview','popularity','poster_path','name','vote_average','vote_count','genre_ids']]
  df=pd.concat([df,df1])
```

```python
df.shape
```

```python
for i in range(1,15):
  response=requests.get(f'https://api.themoviedb.org/3/tv/airing_today?api_key={api_key}&language=en-US&page={i}')
  df2=pd.DataFrame(response.json()['results'])
  df2=df2[['id','adult','overview','popularity','poster_path','name','genre_ids','vote_average','vote_count']]
  df=pd.concat([df,df2])
```

```python
df.shape
```

```python
for i in range(1,112):
  response=requests.get(f'https://api.themoviedb.org/3/tv/top_rated?api_key={api_key}&language=en-US&page={i}')
  df3=pd.DataFrame(response.json()['results'])
  df3=df3[['id','adult','overview','popularity','poster_path','name','genre_ids','vote_average','vote_count']]
  df=pd.concat([df,df3])
```
